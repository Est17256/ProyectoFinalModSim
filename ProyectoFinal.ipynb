```python
import random
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Velocidad
def velocidadC1():
    cond= random.randint(20,120)
    if cond<=70:
        return False
    else:
        return True
def velocidadC2():
    cond= random.randint(20,80)
    if cond<=60:
        return False
    else:
        return True
def velocidadC3():
    cond= random.randint(10,60)
    if cond<=40:
        return False
    else:
        return True
```

```python
arrx=[]
def trayecto():
    cam=[]
    sumt=0
    tray=random.randint(1,2)
    if tray==1:
        cam.append(1)
        cam.append(0)
        cam.append(0)
        return 0, cam
    elif tray==2:
        cam.append(2)
        if velocidadC1():
            sumt+=300
        temp=random.randint(1,2)
        if temp==1:
            cam.append(3)
            if velocidadC2():
                sumt+=300
        elif temp==2:
            cam.append(4)
        cam.append(5)
        temp2=random.randint(1,24)
        if 9>=temp2<=12:
            if velocidadC3():
                sumt+=300
    return sumt,cam
sumt,cam=trayecto()
for i in cam:
    arrx.append(i)
arrx.append(sumt)
# print(sumt,cam)
print(arrx)
```

```python
def simulacion(tiempo):
    arr1=[]
    arr2=[]
    arr3=[]
#     arrx=[]
    arr4=[]
    for i in range(tiempo):
        res,x=trayecto()
        arr1.append(res)
        arr2.append(i+1)
        arr3.append(x)
        arrx=[]
        for i in x:
            arrx.append(i)
        arrx.append(res)
        arr4.append(arrx)
    plt.plot(arr2,arr1)
    return sum(arr1),arr3,arr4
```

```python
def masPasado(x):
    prueba=[]
    valor=[]
    for i in x:
        if i not in prueba:
            prueba.append(i)
            valor.append(x.count(i))
    a=max(valor)
    b=valor.index(a)
    return prueba[b],a
```

```python
autos=input("Ingrese la cantidad de autos que quiere simular")
viajes=input("Ingrese la cantidad de viajes realizados")
dfs=[]
for i in range(int(autos)):
    total,x,tot=simulacion(int(viajes))
    caminoM1,caminoM2= masPasado(x)
    dfs.append(tot)
    print("El total de multas del auto:",i+1,"En el intervalo de :",viajes,"Fue de: ",total)
    print("El camino mas utilizado por el auto",i+1,"Fue",caminoM1,"siendo la cantidad:",caminoM2)
```

```python
dataM=[]
# print(dfs)
for i in dfs:
    for j in i:
        print(j)
        dataM.append(j)
```

```python
df=pd.DataFrame(dataM)
df.columns=["trazoA", "trazoB", "trazoC","ValorMulta"]
df=df.fillna(0)
```
